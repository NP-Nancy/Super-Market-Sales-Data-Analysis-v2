```python
data.drop_duplicates(inplace=True)  # This will delete the duplicates
```

```python
if 'Total' in data.columns:
    branch_sales = data.groupby('Branch')['Total'].sum().sort_values(ascending=False)
else:
    print("The 'Total' column is not present in the data.")
```

```python
data['Total'] = data['Quantity'] * data['Unit price']
```

```python
data.head()
```

```python
branch_sales = data.groupby('Branch')['Total'].sum().sort_values(ascending=False) #Top branch by sales
```

```python
import seaborn as sns                    # Sales distribution by payment method
import matplotlib.pyplot as plt

sns.countplot(data, x='Payment')
plt.title("Payment Method Distribution")
plt.show()
```

```python
# Convert Date column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Group by month and calculate total sales
monthly_sales = data.groupby(data['Date'].dt.month)['Total'].sum()

monthly_sales.plot(kind='bar', title="Monthly Sales Trend", ylabel="Total Sales")

```

```python
data.groupby('Product line')['Total'].sum().sort_values(ascending=False)   #Top selling products

```

```python
#Creating Visualization
#1.Payment method distribution
# Bar plot for payment methods
payment_counts = data['Payment'].value_counts()
payment_counts.plot(kind='bar', color=['skyblue', 'orange', 'green'], title='Payment Method Distribution')
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

```python
#2.Sales by product line
product_sales = data.groupby('Product line')['Total'].sum()
product_sales.plot(kind='bar', color='brown', title='Total Sales by Product Line')
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.show()
```

```python
#3.Month;y sales trend
# Convert Date column to datetime
data['Date'] = pd.to_datetime(data['Date'])
data['Month'] = data['Date'].dt.month
# Grouping total sales by month
monthly_sales = data.groupby('Month')['Total'].sum()

# Line plot for monthly sales
monthly_sales.plot(kind='line', marker='o', color='purple', title='Monthly Sales Trend')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```
